#### La data en les columnes c4 i c5 representada el temps en dues part, separem la data i l'hora en una nova columna anomenada time i eliminem la columna de zona horaria. També eliminem les columnes c4 i c5 ####

```python
import requests
import json
```

```python
# URL de la API
api_url = "http://ip-api.com/json/"
```

```python
#lista ips
data_ips=(data['ip'].value_counts(sort=False)).index
```

```python
data_ips
```

```python
data_ips[0]
```

```python
# Definimos los parametros de respuesta que queremos obtener
parametros = 'status,country,countryCode,region,regionName,city,zip,lat,lon,timezone,isp,org,as,query'
data_Ip = {"fields":parametros}
```

```python
def ip_scraping(ip=""):
 # Nos conectamos con la API
 res = requests.get(api_url+ip, data=data_Ip)
 # Obtenemos y procesamos la respuesta JSON
 api_json_res = json.loads(res.content)
 return api_json_res
```

```python
ip_response = ip_scraping(data_ips[0])
ip_response
```

```python
ip_df=pd.DataFrame([ip_scraping(data_ips[0])])
```

```python
ip_df
```

```python
data_ips1 = data_ips[:50]
```

```python
for ip in data_ips1:
    ip_df = ip_df.append([ip_scraping(ip)],ignore_index=True)
```
